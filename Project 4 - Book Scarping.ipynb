```python
categories = {}

for cat in cats :
    
    name = cat.find('a').text.strip()
    
    cat_url = cat.find('a')['href'].replace('..' , 'https://books.toscrape.com/catalogue/category')
    
    
    categories[name] = cat_url
```

```python
def scrape(url):
    
    
    req = requests.get(url)
    con = req.content
    soup = BeautifulSoup(con , 'html.parser')
    
    
    section = soup.find('section')
    ol = section.find('ol')
    lis = ol.find_all('li')
    
    names , rates , prices , stock , books_url = [] , [] , [] , [] , []
    
    for li in lis:
    
        book_name = li.find('div' , attrs = {'class' : 'image_container'}).find('a').find('img')['alt']
        names.append(book_name)
#         print(book_name)

        book_rate = li.find('p' , attrs = {'class' : 'star-rating'})['class'][1]
        rates.append(book_rate)
#         print(book_rate)

        book_price = float(li.find('p' , attrs = {'class' : 'price_color'}).text[1:])
        prices.append(book_price)
#         print(book_price)

        ava_stock = li.find('p' , attrs = {'class' : 'instock availability'}).text.strip()
        stock.append(ava_stock)
#         print(ava_stock)

        book_url = 'https://books.toscrape.com/catalogue/' + li.find('h3').find('a')['href'].replace('../' , '')
        books_url.append(book_url)
#         print(book_url)


#         print('*'*30) 
    
    
    next_pg = ''
    try:
        next_pg = soup.find('ul' , attrs = {'class' : 'pager'}).find('li' , attrs = {'class' : 'next'}).find('a')['href']
    
    except:
        pass
    
    return names , rates , prices , stock , books_url , next_pg
```

```python
def StoreData(cat , names , rates , prices , stock , books_url):
    
    
    m = 'w' if not os.path.exists('Books Data.csv') else 'a'
    
    with open('Books Data.csv' , mode = m , newline='' , encoding='utf-8') as f :
    
    
        writer = csv.DictWriter(f , fieldnames=[
            'Category',
            'Book Name',
            'Book Rate',
            'Book Price(£)',
            'Book Availability',
            'Book Url'
        ])
        
        if m == 'w':
        
            writer.writeheader()
            
        
        for book_name , book_price , book_rate , ava_stock , book_url in zip(names , rates , prices , stock , books_url):
            
            writer.writerow({
                'Category' : cat,    
                'Book Name' : book_name,
                'Book Price(£)' : book_price,
                'Book Rate' : book_rate,
                'Book Availability' : ava_stock,
                'Book Url' : book_url
                    })
```

```python
import time
```

```python
for cat , url in categories.items():
    
    
    # scrape first page for each category
    print(url)
    names , rates , prices , stock , books_url , next_pg = scrape(url)
    
    # store data
    StoreData(cat , names , rates , prices , stock , books_url)
    
    time.sleep(0.5)
    # scrape for all pages after first page 
    while next_pg != '':
        
        
        new_url = url.replace('index.html' , next_pg)
        print(new_url)
        
        names , rates , prices , stock , books_url , next_pg = scrape(new_url)
        
        StoreData(cat , names , rates , prices , stock , books_url)
        
        time.sleep(0.5)
        
    print('*'*30)
```
